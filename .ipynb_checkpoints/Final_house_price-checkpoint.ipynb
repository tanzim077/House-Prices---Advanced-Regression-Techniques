```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('./my_train.csv')
test = pd.read_csv('./my_test.csv')
test1 = pd.read_csv('Dataset/test.csv')
```

```python
train.shape
```

```python
test.shape
```

```python
all_data = pd.concat([train,test])
```

```python
all_data.shape
```

```python
all_data.to_csv('all.csv', index=False)
```

### Seperate categorical feature for encoding

```python
col = all_data.columns
```

```python
cat_col = []

for item in col:
    if all_data[item].dtype == 'object':
        cat_col.append(item)
```

```python
len(cat_col)
```

### Make a function for encode all the categorical columns

```python
dummies = pd.get_dummies(all_data[cat_col])
```

```python
dummies.shape
```

```python
final_data = all_data
```

```python
final_data.drop(cat_col,axis=1,inplace=True)
```

```python
final_data.shape
```

```python
final_data = pd.concat([final_data,dummies], axis=1)
```

```python
final_data.shape
```

```python
final_data = final_data.loc[:,~final_data.columns.duplicated()]
```

```python
final_data.shape
```

```python
x = final_data.loc[:1459]
```

```python
y = final_data[1460:]
```

```python
x.shape
```

```python
y.shape
```

```python
x['SalePrice'].isnull().sum()
```

```python
xtrain = x.drop(['SalePrice'], axis=1)
ytrain = x['SalePrice']
xtest = y
```
