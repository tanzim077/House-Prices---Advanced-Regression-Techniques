```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('./Dataset/train.csv')
test = pd.read_csv('./Dataset/test.csv', index_col='Id')
```

```python
plt.figure(figsize=(25,5)) # set figure size 20 width 5 height
plt.subplot(1,2,1) # create subplot 1 rows, 2 columns and 1 no position
sns.heatmap(train.isnull(),yticklabels=False,cbar=False);
plt.title('train');
plt.subplot(1,2,2) # create subplot 1 rows, 2 columns and 2 no position
sns.heatmap(test.isnull(),yticklabels=False,cbar=False);
plt.title('test');
```

```python
def null_check(x):
    null = []
    c = x.columns
    for i in c:
        if x[i].isnull().sum() > 0:
            null.append(i)
    
    return null
```

```python
null_o =  null_check(ndf_o)
null_n =  null_check(ndf_n)
```

```python
null_o
```

```python
for i in null_o:
    print(i)
    print(ndf_o[i].isnull().sum())
```

```python
drop_columns = ['Alley','PoolQC','FireplaceQu','Fence','MiscFeature']
```

```python
ndf_o.drop(drop_columns,axis=1,inplace=True)
```

```python
def fill_by_mean(x):
    if len(x) > 0:
        for i in x:
            x[i].fillna((x[i].mean()), inplace=True)
            
def fill_by_mode(x):
    if len(x) > 0:
        for i in x:
            x[i].fillna((x[i].mode()[0]), inplace=True)
```

```python
fill_by_mean(ndf_n)
fill_by_mode(ndf_o)
```
