## Null Value Check

```python
col = test.columns
valid_col = []
null_col= []

for i in col:
    x = test[i].isnull().sum()
    if x > 0:
        null_col.append(i)
    else:
        valid_col.append(i)
```

```python
print("Total rows : ", len(test))
print("Non Null Columns : ", len(valid_col))
print("Null columns : ",  len(null_col))
```

```python
null_col
```

```python
for i in null_col:
    print(i, "=",  test[i].isnull().sum())
```

```python
sns.heatmap(test.isnull(),yticklabels=False,cbar=False);
plt.title('test');
```

## Seperate columns by their datatype

```python
def df_null_check(item):
    col = item.columns
    intfloatdt = []
    objectdt = []
    for i in col:
        x = item[i].isnull().sum()
        print(i, '=', x)
```

```python
df_null_check(test)
```

```python
col = test.columns
intfloatdt = []
objectdt = []
for item in col:
    x = test[item].isnull().sum()
    if x > 0:
        dt = test[item].dtypes
        if dt == 'int64' or dt == 'float64':
            intfloatdt.append(item)
        elif dt == 'object':
            objectdt.append(item)
```

```python
intfloatdt
```

```python
objectdt
```

```python
for item in  intfloatdt:
    print(item , '=' , test[item].isnull().sum())
```

```python
for item in intfloatdt:
    print('\t', item, '\n')
    print(test[item].describe(), "\n")
```

```python
for item in  objectdt:
    print(item , '=' , test[item].isnull().sum())
```

```python
for item in objectdt:
    print('\t', item, '\n')
    print(test[item].describe())
```

```python
print("Null Int and float = ", len(intfloatdt))
print("Null Object type = ", len(objectdt))
```

## Fill Missing Values

**Fill int and float type missing value using mean**  
**Fill object type missing value using mode**

```python
for item in intfloatdt:
    test[item]=test[item].fillna(test[item].mean())
    
for item in objectdt:
    test[item]=test[item].fillna(test[item].mode()[0])    
```

```python
df_null_check(test)
```
